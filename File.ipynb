```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove USA from event name 
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#clean up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split('').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns: Athlete club, Athlete countty# athlete year of birth, Athlete Age category
```

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
#clean up  null values 
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2[df2.duplicated() == True]
```

```python
df2.reset_index(drop = True)
```

```python
df2['athlete_id'] = df2['athlete_id'].astype(int)
```

```python
#fix types 
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete ID'] = df2['Athlete ID'].astype(int)
```

```python
df2.head(5)
```

```python
df2['athlete_average_speed'] = df2['athlete_average_speed'].astype(float)
```

```python
df2['race_number_of_finishers'] = df2['race_number_of_finishers'].astype(int)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#rename columns 
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers    float64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                   float64
#athlete_age                    int32
#dtype: object
df2 = df2.rename(columns = {'Year of event':'year' ,
                            'Event dates':  'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performance',
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID': 'athlete_id'
                           })
                            
                             
```

```python
df2.head()
```

```python
#Difference in speed for 50k,50mi male to female
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50miles Race(20 + race min)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = False).query('count>19').head(15)
```

```python
#what age groups and the worst in the 50 mi Race(20+races min)(show 20)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = True).query('count>9').head(20)
```
